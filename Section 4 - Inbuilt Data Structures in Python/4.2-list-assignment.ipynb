```python
### Assignment 6: List Methods

## Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

## Ascending

nums = [2,66,12,990,1,11,45, 66]
nums.sort()
print(nums)  # [1, 2, 11, 12, 45, 66, 66, 990]

## Descending

print(nums[::-1])  # [990, 66, 66, 45, 12, 11, 2, 1]

## Removing Duplicates

unique = list(set(nums))  # We use the set method to remove duplicates
print(unique)  # [1, 2, 66, 11, 12, 45, 990]
```

```python
### Assignment 8: List of Dictionaries

## Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

marks_list = [
    {"name": "Nishank", "score": 89},
    {"name": "Koul", "score": 92},
    {"name": "John", "score": 77},
    {"name": "Jacob", "score": 12}
]

marks_list.sort(key=lambda x: x['score'], reverse=True)
print(marks_list)

'''
[{'name': 'Koul', 'score': 92}, {'name': 'Nishank', 'score': 89}, {'name': 'John', 'score': 77}, {'name': 'Jacob', 'score': 12}]
'''
```

```python
### Assignment 9: Matrix Transposition

## Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

def transpose_matrix(matrix):
    # Using list comprehension
    transposed = [[matrix[j][i] for j in range(3)] for i in range(3)]
    return transposed

# Example 3x3 matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Original Matrix:")
for row in matrix:
    print(row)

# Call the function here ðŸ‘‡
transposed = transpose_matrix(matrix)

print("\nTransposed Matrix:")
for row in transposed:
    print(row)
```

```python
### Assignment 10: Flattening a Nested List

## Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

def flatten_list(nested_lst):
    # Create a new list
    flat = []
    for sublist in nested_lst:
        for item in sublist:
            flat.append(item)
    return flat

# Example nested list
nested_lst = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

print("Original Nested List:")
print(nested_lst)

flattened = flatten_list(nested_lst)

print("\nFlattened List:")
print(flattened)
```

```python
### Assignment 11: List Manipulation

## Create a list of the first 10 positive integers. Remove the elements at indices 2, 4, and 6, and insert the element '99' at index 5. Print the modified list.

lst_10 = [x for x in range(1,11)]
print(lst_10) 
lst_10.pop(2)
lst_10.pop(4)
lst_10.pop(6)
lst_10.insert(5, 99)
print(lst_10)  # [1, 2, 4, 5, 7, 99, 8, 10]
```

```python
### Assignment 12: List Zipping

## Create two lists of the same length. Use the `zip` function to combine these lists into a list of tuples and print the result.

nums1 = [1,2,3,4,5]
nums2 = [6,7,8,9,10]

# Combine the lists using zip
zipped_list = list(zip(nums1, nums2))

print(zipped_list)
```

```python
### Assignment 13: List Reversal

## Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.

def reverse_lst(original_lst):
    original_lst.reverse()
    return original_lst

original_lst = ["Hello", "in", "this", "world", "Nishank"]

print(f"Original List: {original_lst}")

reversed_list = reverse_lst(original_lst)

print(f"Reversed List: {reversed_list}")
```

```python
### Assignment 14: List Rotation

## Write a function that rotates a list by n positions. Print the original and rotated lists.

def rotate_list(original_list, n):
    # Handle rotation greater than list length
    n = n % len(original_list)
    # Rotate by slicing
    rotated = original_list[-n:] + original_list[:-n]
    return rotated

# Example list
original_list = [1, 2, 3, 4, 5, 6]

print("Original List:")
print(original_list)

rotated = rotate_list(original_list, 2)  # Rotate by 2 positions

print("\nRotated List:")
print(rotated)
```

```python
### Assignment 15: List Intersection

## Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

n1 = [1,2,3,4,5,6,7]
n2 = [4,5,6]

intersected_list = [x for x in n1 if x in n2]

print(intersected_list)
```
