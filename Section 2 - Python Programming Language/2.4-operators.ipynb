```python
## Arithmetic Operators

num1 = 10
num2 = 5

add = num1 + num2
subt = num1 - num2
multi = num1 * num2
divide = num1/num2
floor_divide = num1//num2  ## Floor Division gives you the floor value removing the decimal part.
exponent = num1 ** num2
modulus = num1 % num2

print(add)  ## 15
print(subt)  ## 5
print(multi)  ## 20
print(divide)  ## 2.0
print(floor_divide)  ## 2
print(exponent)  ## 100000
print(modulus)  ## 0
```

```python
## Comparison Operators

## Equal To Operator (==)

a = 10
b = 20

print(a==b)  ## False

## Not Equal To Operator (!=)

print(a!=b)  ## True

## Greater Than Operator (>)

print(a > b)  ## False
print(b > a)  ## True

## Less Than Operator (<)

print(a < b)  ## True
print(b < a)  ## False

## Greater Than or Equal To (>=)

print(a >= b)  ## False
print(b >= a)  ## True

## Less Than or Equal To (<=)

print(a <= b)  ## True
print(b <= a)  ## False
```

```python
## Logical Operators

## AND

X1 = True
Y1= True

print(X1 and Y1)  ## True

X2 = True
Y2 = False

print(X2 and Y2)  ## False

X3 = False
Y3 = True

print(X3 and Y3)  ## False

X4 = False
Y4 = False

print(X4 and Y4)  ## False

## OR

print(X1 or Y1)  ## True
print(X2 or Y2)  ## True
print(X3 or Y3)  ## True
print(X4 or Y4)  ## True

## NOT

X5 = True
print(not X5)  ## False (just inverse)
```
